```python
import json
import requests
import pandas as pd
```

```python
api_key = json.loads(open("../API/YouTube/api_key.txt").read())['api_key']
```

```python
def get_metadata():
    #function retrieves nextpageToken, videoId and title through search list
    
    endpoint = 'https://www.googleapis.com/youtube/v3/search?q={}&videoCategoryId={}&publishedAfter={}&publishedBefore={}&type={}&maxResults={}&fields={}&part={}&key={}'
    #fills brackets w/ appropriate search filters for ad videos
    url = endpoint.format('%23sponsored -non+sponsored -not+sponsored', 
                                    #query URL encode: %23 = #, %7C =|, " -" without
                          #other options: #%23sponsored -non-sponsored -not+sponsored or %23ad%7C%23sponsored -ads
                          26, #videoCategoryId
                          '2018-08-01T00:00:00Z', #publishedAfter
                          '2018-09-01T00:00:00Z', #publishedBefore
                          'video', #type
                          '50', #maxResults
                          'nextPageToken,items(id(videoId)),items(snippet(channelId)), items(snippet(title))', #fields ##GET CHANNEL ID TOO ,items(snippet(title))
                          'snippet', #part
                          api_key) #key
    
    response = requests.get(url).json()
    responses.append(response)
    #sets next_reponse to response
    next_response = response #initialize a variable next_response, set it to response
    return response
```

# Storing Data + Getting Comments

```python
def get_comments(videoId):

    #function that returns comments of a video given the videoId

    endpoint = 'https://www.googleapis.com/youtube/v3/commentThreads?part=snippet&videoId={}&key={}&maxResults={}'
    url = endpoint.format(videoId, api_key, 100)

    response = requests.get(url).json()
#     print(response)
    
    comments = []
    i = 0
    
    if 'items' in response.keys():
        while i < len(response['items']):
            #goes through the items_list and accesses the textOriginal
            comment_text = response['items'][i]['snippet']['topLevelComment']['snippet']['textOriginal']
            #adds to list
            comments.append(comment_text)
            i += 1
    
    return comments
```

```python
def data_dict():
    #make a dictionary of the videoIds
    
    #first make a list of keys (videoIds)
    Ids = []
    metadata = []
    for i in range(len(responses)):
        for j in range(50): #change to 50 when tested
            
            videoId = responses[i]['items'][j]['id']['videoId']
#             print(videoId)
            
            #first number is first pull request, second number is response/50
            Ids.append(videoId)
            #make a lists to make inner dict
            meta_dict_keys = ['title','channel Id', 'comments']
            title = responses[i]['items'][j]['snippet']['title']
            channelId = responses[i]['items'][j]['snippet']['channelId'] 
            
            comments = get_comments(videoId)
            meta_dict_values = [title, channelId, comments]
            
            #intialize dict
            meta_dict = dict(zip(meta_dict_keys, meta_dict_values))
            
            #add to list
            metadata.append(meta_dict) 
            
            j+=1      
        i+=1
        
    #now put them all into a dictionary
    dictionary = dict(zip(Ids, metadata))
    
    with open("../Fellowship_Spring2019/youtube-videos/data_dict.json", 'w') as f:
            json.dump(dictionary, f)
                
    return dictionary
```
