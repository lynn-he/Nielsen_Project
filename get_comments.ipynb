```python
api_key = json.loads(open("../API/YouTube/api_key.txt").read())['api_key']
```

```python
def get_comments(videoId):

    #function that returns comments of a video given the videoId

    endpoint = '''https://www.googleapis.com/youtube/v3/commentThreads?part=snippet&videoId={}&key={}&maxResults=100&fields={}'''
    url = endpoint.format(videoId, api_key, 'items')

    response = json.loads(requests.get(url).text)

    return response   
```
